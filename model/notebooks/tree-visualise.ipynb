# Minh họa Decision Tree

Phần này sẽ dùng sklearn.tree.export_text để in ra cây quyết định

```python
from sklearn.tree import export_text
import pickle as pkl
```

```python
__MODEL__ = '../model/'
```

Load model và vectorizer (chủ yếu là cần bộ features)

```python
with open(__MODEL__ + 'vectorizer.pkl', 'rb+') as f:
    vectorizer = pkl.load(f)

with open(__MODEL__ + 'decision_tree.pkl', 'rb+') as f:
    decision_model = pkl.load(f)
```

```python
tree_rules = export_text(decision_model, feature_names = vectorizer.get_feature_names())
```
